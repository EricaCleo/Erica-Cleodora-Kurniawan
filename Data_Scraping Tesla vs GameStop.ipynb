```python
from bs4 import BeautifulSoup
import pandas as pd

soup = BeautifulSoup(html_data.text, 'html.parser')

# Create an empty list to store data
data = []

# Iterate through each row of the table
for row in soup.find("tbody").find_all("tr"):
    # Extract data from each column (td tag)
    col = row.find_all("td")
    # Extract Date and Revenue from each column
    Date = col[0].text
    Revenue = col[1].text
    # Append the data to the list
    data.append({"Date": Date, "Revenue": Revenue})

gme_revenue = pd.DataFrame(data)
gme_revenue.head()
```

```python
import matplotlib.pyplot as plt

# Data
gme_revenue_data = {
    'Date': ['2016','2017','2018','2019','2020'],
    'Revenue': [9364, 7965, 8547, 8285, 6466]
}

tesla_revenue_data = {
    'Date': ['2016','2017','2018','2019','2020'],
    'Revenue': [11759, 21461, 24578, 31536, 53823]
}

gme_revenue_df = pd.DataFrame(gme_revenue_data)
tesla_revenue_df = pd.DataFrame(tesla_revenue_data)


plt.figure(figsize=(10, 6))

plt.plot(gme_revenue_df['Date'], gme_revenue_df['Revenue'], marker='o', label='GameStop', color='blue')
plt.plot(tesla_revenue_df['Date'], tesla_revenue_df['Revenue'], marker='o', label='Tesla', color='red')

plt.title('Revenue Comparison')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
